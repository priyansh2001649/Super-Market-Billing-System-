```python
# Dictionary of available supermarket items and their prices
supermarket_items = {
    "Rice": 50,
    "Sugar": 45,
    "Milk": 60,
    "Bread": 40,
    "Eggs (Dozen)": 70,
    "Oil (1L)": 150,
    "Salt": 20,
    "Flour": 35,
    "Biscuits": 25,
    "Soap": 30,
    "Shampoo": 120,
    "Toothpaste": 80,
}

# Main loop to handle multiple customers
while True:
    print("\nWelcome to the Supermarket! Let's help you with your shopping.")
    customer_name = input("Enter your name: ").strip()
    contact_number = input("Enter your phone number: ")

    # Validate phone number
    while len(contact_number) != 10 or not contact_number.isdigit():
        print("Invalid phone number. Please enter a valid 10-digit number.")
        contact_number = input("Enter your phone number: ")

    # Display product list if requested
    display_list = input("Would you like to see the product list? (Yes/No): ").capitalize()
    if display_list == "Yes":
        print("\nAvailable Items:")
        for item, price in supermarket_items.items():
            print(f"{item:<15} - Rs. {price}")
        print("\nLet's start adding items to your cart!")

    # Initialize variables
    total_amount = 0
    cart = {}

    # Adding items to the cart
    while True:
        product = input("Enter the product you want to buy: ").title()
        if product in supermarket_items:
            quantity = int(input(f"How many units of {product} would you like to buy? "))
            cart[product] = [supermarket_items[product], quantity]
            total_amount += supermarket_items[product] * quantity
            print(f"Current Total: Rs. {total_amount}")
        else:
            print(f"Sorry, {product} is not available. Please choose from the product list.")

        add_more = input("Do you want to add more items? (Yes/No): ").capitalize()
        if add_more == "No":
            break

    # Apply discounts based on total amount
    if total_amount >= 5000:
        discount = total_amount * 0.20
        discount_msg = "20% Discount Applied"
    elif total_amount >= 3000:
        discount = total_amount * 0.15
        discount_msg = "15% Discount Applied"
    elif total_amount >= 1000:
        discount = total_amount * 0.10
        discount_msg = "10% Discount Applied"
    else:
        discount = 0
        discount_msg = "No Discount Applied"

    # Payment method and card discount
    payment_mode = input("Enter payment method (Cash/Card): ").capitalize()
    if payment_mode == "Card":
        card_discount = 50
        print("Rs. 50 discount for card payment applied!")
    else:
        card_discount = 0

    # Generating the receipt
    print("\n" + "*" * 40)
    print(f"{'-'*10} Thank You for Shopping {'-'*10}")
    print(f"Customer Name: {customer_name}")
    print(f"Contact Number: {contact_number}")
    print("-" * 40)
    print(f"{'Item':<15}{'Price':<8}{'Qty':<6}{'Total':<10}")
    print("-" * 40)
    for item, details in cart.items():
        print(f"{item:<15}{details[0]:<8}{details[1]:<6}{details[0]*details[1]:<10}")
    print("-" * 40)
    print(f"Subtotal: Rs. {total_amount}")
    print(f"{discount_msg}: Rs. {discount:.2f}")
    if payment_mode == "Card":
        print(f"Card Discount: Rs. {card_discount}")
    print("=" * 40)
    print(f"Payable Amount: Rs. {total_amount - discount - card_discount}")
    print("=" * 40)
    print("*" * 40 + "\n")

    # Check if a new customer needs to be processed
    next_customer = input("Do you want to process another customer? (Yes/No): ").capitalize()
    if next_customer == "No":
        print("Thank you for visiting the Supermarket! Have a great day!")
        break
```
